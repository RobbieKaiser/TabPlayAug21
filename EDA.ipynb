```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
train = pd.read_csv('Data/train.csv')
train.head()
```

```python
train.describe()
```

```python
#cols to investigate distributions skew
#f1, f2,
```

```python
features = train.loc[:, train.columns.str.contains('f')]
features.head()
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
```

```python
quant_array = [0.01, 0.05, 0.1, 0.5, 0.9, 0.95, 0.99]
```

```python
for col in features.columns:
    plt.hist(features[col], bins = 30)
    
    quantiles = np.quantile(features[col], quant_array)
    quant_idx = 0
    for val in quantiles:
        plt.axvline(x=val, c = 'red')
        plt.text(val,1000,quant_array[quant_idx],rotation=90)
        quant_idx = quant_idx+1
    
    plt.title(col)
    plt.show()
```
